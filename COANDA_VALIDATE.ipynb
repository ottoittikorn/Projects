```python


# 3. Scatter plot (dot at each position, colored by direction, with custom colors and legend)
direction_map = {
    "down": -1,
    "none": 0,
    "both": 1,
    "up": 2,
}

df["direction_code"] = df["direction"].map(direction_map)
df2["direction_code"] = df2["direction"].map(direction_map)

palette = sns.color_palette("tab20",20)  # 4 unique colors

# direction_colors = {
#     -1: "red",    # down
#     0: "grey",    # none
#     1: "blue",    # both
#     2: "green",   # up
# }

direction_colors = {
    -1: palette[0],    # down
    0: palette[8],     # none
    1: palette[3],     # both
    2: palette[4],     # up
}

direction_labels = {
    -1: "Downwards",
    # -1.5: "Mostly downwards",
    0: "No Coanda",
    1: "Both ways",
    # 1.5: "Mostly upwards",
    2: "Upwards",
}

direction_markers = {
    -1: "v",      # Downwards: triangle down
    # -1.5: "v",    # Mostly downwards: triangle down
    0: "o",       # No Coanda: circle
    1: "D",       # Both ways: diamond
    # 1.5: "^",     # Mostly upwards: triangle up
    2: "^",       # Upwards: triangle up
}


plt.figure(figsize=(6, 4))
for code, color in direction_colors.items():
    subset = df[df["direction_code"] == code]
    marker_style = direction_markers.get(code, "o")  # Default to circle if code not found
    plt.scatter(
        subset["salt_conc"],
        subset["flow_rate"],
        color=color,
        s=120,
        edgecolor="k",
        alpha=0.8,
        label=direction_labels[code],
        marker=marker_style
    )
plt.ylabel("Flowrate (ml/min)")
plt.xlabel("Salt Concentration (wt%)")
plt.legend(title="Direction")
plt.tight_layout()          
plt.show()

# # ----------------------------
# # 4. Direction distribution plot
# # ----------------------------
# plt.figure(figsize=(8, 5))
# sns.countplot(data=df, x="flow_rate", hue="direction")
# plt.title("Direction Distribution per Flow Rate")
# plt.tight_layout()
# plt.show()
```

```python
def clean_direction(label):
    if label in ["down", "down bit up"]:
        return "down"
    elif label in ["up", "up bit down"]:
        return "up"
    elif label in ["none"]:
        return "none"
    elif label in ["both"]:
        return "both"
    else:
        return "-"

df["direction_clean"] = df["direction"].apply(clean_direction)
df2["direction_clean"] = df2["direction"].apply(clean_direction)


df = df.rename(columns={"direction_clean": "direction_exp"})
df2 = df2.rename(columns={"direction_clean": "direction_cfd"})
```

```python
df_merged = pd.merge(
    df[["flow_rate", "salt_conc", "direction_exp"]],
    df2[["flow_rate", "salt_conc", "direction_cfd"]],
    on=["flow_rate", "salt_conc"],
    how="inner"
)

df_merged["match"] = (
    df_merged["direction_exp"] == df_merged["direction_cfd"]
)

accuracy = df_merged["match"].mean()
print("Accuracy:", accuracy)

```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

labels = ["down", "none", "up"]

cm = confusion_matrix(
    df_merged["direction_exp"],
    df_merged["direction_cfd"],
    labels=labels
)

# Convert to DataFrame for labels
cm_df = pd.DataFrame(cm, index=labels, columns=labels)

# Plot
plt.figure(figsize=(6,5))
sns.heatmap(cm_df, annot=True, fmt="d", cmap="Blues")

plt.xlabel("CFD Prediction")
plt.ylabel("Experimental")
plt.title("Confusion Matrix: CFD vs Experimental Coandă Direction")

plt.tight_layout()
plt.show()
```

```python

# Richardson Number vs Flowrate
import numpy as np
import matplotlib.ticker as ticker

# Constants
g = 9.81  # m/s^2
area = 0.0016 * 0.003  # m^2 (1.6mm x 3mm)
L = 0.0016 #242/1e9  # m (characteristic length) Latex length

# Assign densities based on salt concentration
def get_density(salt_conc):
    # Use measured values for 2.43 and 10, otherwise linear interpolation
    if (salt_conc == 2.43):
        return 1.01417 * 1000
    elif (salt_conc == 10):
        return 1.06888 * 1000
    rho_water = 998
    rho_salt = 1.00409 * 1000   # example: NaCl
    rho_solution = (rho_salt - rho_water) * salt_conc + rho_water
    return rho_solution

# Calculate velocity (U) and Richardson number for each row
def compute_richardson(row):
    # if row["direction"] == "-":
    #     return np.nan
    rho1 = get_density(row["salt_conc"])  # Use the function for density
    rho2 = 1001  # Reference density (could be water or another phase)
    # rho_avg = (rho1 + rho2) / 2
    Q = row["flow_rate"] / 1e6 / 60
    U = Q / area
    # if U == 0:
    #     return np.nan
    Ri = g * abs((rho1 - rho2) * L / (rho2 * U**2))
    return Ri

df["Richardson"] = df.apply(compute_richardson, axis=1)

# Plot Richardson number (y) vs flowrate (x)
plt.figure(figsize=(12, 6))
for code, color in direction_colors.items():
    subset = df[(df["direction_code"] == code) & (~df["Richardson"].isna())]
    marker_style = direction_markers.get(code, "o")
    plt.scatter(
        subset["flow_rate"],
        subset["Richardson"],
        color=color,
        s=120,
        edgecolor="k",
        alpha=0.8,
        label=direction_labels[code],
        marker=marker_style
    )
plt.xlabel("Flowrate (ml/min)")
plt.ylabel("Richardson Number")
plt.yscale("log")
plt.axhline(y=1, color='black', linestyle='--', linewidth=1.5, label='Ri = 1')
plt.xscale("log")

flowrates = sorted(df["flow_rate"].unique())
plt.xticks(flowrates, [str(fr) for fr in flowrates])

# Set y-limits and ticks: adjust these as needed for your data
ymin, ymax = 0.1, 1e4  # Example: from 0.1 to 10,000
plt.ylim(ymin, ymax)

# Get current yticks, replace 1 with '1', others as log
ax = plt.gca()
yticks = ax.get_yticks()
labels = []
for y in yticks:
    if y == 1:
        labels.append("1")
    elif y > 0:
        labels.append(f"$10^{{{int(np.log10(y))}}}$")
    else:
        labels.append("")
ax.set_yticks(yticks)
ax.set_yticklabels(labels)

plt.legend(title="Direction")
plt.tight_layout()
plt.show()
```
