```python
def clean_direction(label):
    if label in ["down", "down bit up"]:
        return "down"
    elif label in ["up", "up bit down"]:
        return "up"
    elif label in ["none"]:
        return "none"
    elif label in ["both"]:
        return "both"
    else:
        return "-"

df["direction_clean"] = df["direction"].apply(clean_direction)
df2["direction_clean"] = df2["direction"].apply(clean_direction)


df = df.rename(columns={"direction_clean": "direction_exp"})
df2 = df2.rename(columns={"direction_clean": "direction_cfd"})
```

```python
df_merged = pd.merge(
    df[["flow_rate", "salt_conc", "direction_exp"]],
    df2[["flow_rate", "salt_conc", "direction_cfd"]],
    on=["flow_rate", "salt_conc"],
    how="inner"
)

df_merged["match"] = (
    df_merged["direction_exp"] == df_merged["direction_cfd"]
)

accuracy = df_merged["match"].mean()
print("Accuracy:", accuracy)

```

```python

# Richardson Number vs Flowrate
import numpy as np
import matplotlib.ticker as ticker

# Constants
g = 9.81  # m/s^2
area = 0.0016 * 0.003  # m^2 (1.6mm x 3mm)
L = 0.0016 #242/1e9  # m (characteristic length) Latex length

# Assign densities based on salt concentration
def get_density(salt_conc):
    # Use measured values for 2.43 and 10, otherwise linear interpolation
    if (salt_conc == 2.43):
        return 1.01417 * 1000
    elif (salt_conc == 10):
        return 1.06888 * 1000
    rho_water = 998
    rho_salt = 1.00409 * 1000   # example: NaCl
    rho_solution = (rho_salt - rho_water) * salt_conc + rho_water
    return rho_solution

# Calculate velocity (U) and Richardson number for each row
def compute_richardson(row):
    # if row["direction"] == "-":
    #     return np.nan
    rho1 = get_density(row["salt_conc"])  # Use the function for density
    rho2 = 1001  # Reference density (could be water or another phase)
    # rho_avg = (rho1 + rho2) / 2
    Q = row["flow_rate"] / 1e6 / 60
    U = Q / area
    # if U == 0:
    #     return np.nan
    Ri = g * abs((rho1 - rho2) * L / (rho2 * U**2)) # density diffrence always positive 
    return Ri

df["Richardson"] = df.apply(compute_richardson, axis=1)

# Plot Richardson number (y) vs flowrate (x)
plt.figure(figsize=(12, 6))
for code, color in direction_colors.items():
    subset = df[(df["direction_code"] == code) & (~df["Richardson"].isna())]
    marker_style = direction_markers.get(code, "o")
    plt.scatter(
        subset["flow_rate"],
        subset["Richardson"],
        color=color,
        s=120,
        edgecolor="k",
        alpha=0.8,
        label=direction_labels[code],
        marker=marker_style
    )
plt.xlabel("Flowrate (ml/min)")
plt.ylabel("Richardson Number")
plt.yscale("log")
plt.axhline(y=1, color='black', linestyle='--', linewidth=1.5, label='Ri = 1')
plt.xscale("log")

flowrates = sorted(df["flow_rate"].unique())
plt.xticks(flowrates, [str(fr) for fr in flowrates])

# Set y-limits and ticks: adjust these as needed for your data
ymin, ymax = 0.1, 1e4  # Example: from 0.1 to 10,000
plt.ylim(ymin, ymax)

# Get current yticks, replace 1 with '1', others as log
ax = plt.gca()
yticks = ax.get_yticks()
labels = []
for y in yticks:
    if y == 1:
        labels.append("1")
    elif y > 0:
        labels.append(f"$10^{{{int(np.log10(y))}}}$")
    else:
        labels.append("")
ax.set_yticks(yticks)
ax.set_yticklabels(labels)

plt.legend(title="Direction")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.ticker as ticker

# Plotting v^2 vs Delta rho / rho

Ri = 1  # Richardson number at the transition point

m = g*0.0016/Ri

# Calculate velocity (U) and Richardson number for each row
def compute_v_2(row):
    if row["direction"] == "-":
        return np.nan
    rho1 = get_density(row["salt_conc"])  # Use the function for density
    rho2 = 1001  # Reference density (could be water or another phase)
    # rho_avg = (rho1 + rho2) / 2
    Del_rho = abs((rho1 - rho2)/rho2)
    Q = row["flow_rate"] / 1e6 / 60
    U = Q / area
    if U == 0:
        return np.nan
    U_s = U**2
    return U_s, Del_rho

df2[["v_2", "Del_rho"]] = pd.DataFrame(df.apply(compute_v_2, axis=1).tolist(), index=df2.index)

# Plot Richardson number (y) vs flowrate (x)
plt.figure(figsize=(10, 6))
for code, color in direction_colors.items():
    subset = df2[(df2["direction_code"] == code) & (~df2["v_2"].isna()) & (~df2["Del_rho"].isna())]
    marker_style = direction_markers.get(code, "o")
    plt.scatter(
        subset["Del_rho"],
        subset["v_2"],
        color=color,
        s=120,
        edgecolor="k",
        alpha=0.8,
        label=direction_labels[code],
        marker=marker_style
    )
plt.xlabel("Δρ / ρ")
plt.ylabel("v² (m²/s²)")
plt.yscale("log")
plt.xscale("log", base=10, subs=[1, 2, 5])  # Log scale with custom ticks
# plt.axline((0, 0), slope=m, color='black', linestyle='--', linewidth=2, label='Ri = 1')
plt.gca().xaxis.set_major_formatter(ticker.FormatStrFormatter('%.3f'))

# Get the range of x (Del_rho) values
x_min = df2["Del_rho"].min()
x_max = df2["Del_rho"].max()
x_vals = np.logspace(np.log10(x_min), np.log10(x_max), 100)
y_vals = m * x_vals
plt.plot(x_vals, y_vals, color='black', linestyle='--', linewidth=1.5, label='Ri = 1')
plt.legend(title="Direction")
plt.tight_layout()
plt.show()
```
